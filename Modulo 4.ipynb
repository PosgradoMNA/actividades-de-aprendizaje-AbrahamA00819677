```python
#Modes "r" Read mode for reading files
# "w" Write mode for writing files
```

```python
# Read the Example1.txt
example1 = "Example1.txt"
file1 = open(example1, "r")
```

```python
# Print the path of file

file1.name
```

```python
# Print the mode of file, either 'r' or 'w'

file1.mode
```

```python
# Read the file

FileContent = file1.read()
FileContent
```

```python
# Print the file with '\n' as a new line

print(FileContent)
```

```python
# Type of file content

type(FileContent)
```

```python
# Es importante que el archivo se cierre al final de uso con el objetivo de ahorrar espacio.
# Close file after finish

file1.close()
```

```python
# Una mejor forma de abrir un archivo
# Usando el comando "with" atomaticamente cierra el archivo.

#Open file using with

with open(example1, "r") as file1:
    FileContent = file1.read()
    print(FileContent)
```

```python
# Verify if the file is closed

file1.closed
```

```python
# See the content of file

print(FileContent)
```
